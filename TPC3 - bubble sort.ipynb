```python
def menu():
    print("Gestao de numeros inteiros")
    print("(1) Introduza a lista a mao")
    print("(2) Gerar lista com valores aleatorios")
    print("(3) Bubble sort")
    print("(0) Sair")
    
    op = input("Qual e a sua opcao: ")
    
    if op == "1":
            lis = CriarListaM()
            
            print(lis)
            print(""
                  
                  "")
            menu()
            
    elif op =="2":
            lis2 = CriarListaA()
            
            
            print(lis2)
            print(""
                  
                  "")
            menu()
            
            
        
    elif op == "3":
        num = input("utilizar (1)lista a mao ou (2)lista aleatoria?:")
        if num == "1":
            lista = CriarListaM()
            lis3 = bubblesort(lista)
            print(lis3)
            menu()
            
            
        elif num == "2":
            lista = CriarListaA()
            lis3 = bubblesort(lista)
            print(lis3)
            menu()
            
        else:
            print("opcao nao reconhecida")
            menu()
            
        lis3 = bubblesort(lista)
        print(lis3)
        menu()
        
        
    else:
            print("Esta opcao nao existe")
            menu()
            
    if op == "0":
            exit()
    
    
def CriarListaM():
    n = input("Introduza o numero de elementos da lista:")
    n = int(n)
    lista = []
    i = 0
    while i != n :
        elemento = input("Introduza o elemento " + str(i + 1) + ":")
        lista.append(elemento)
        i = i + 1
    return(lista)
    
def CriarListaA ():
    import random
    n = int(input("Introduza o numero de elementos da lista:"))
    lista2 = []
    for i in range(n):
        elemento = random.randint(0,100)
        lista2.append(elemento)
    return(lista2)
    
def bubblesort(lista):
    
        
    for i in lista:
        n = lista.index(i) 
        if n == len(lista) - 1 :
            return lista
            
    
        elif int(i) > int(lista[n +1]):
    
            lista.pop(n)
            lista.insert(n + 1 , i)
            bubblesort(lista)
        
        
menu()
```
