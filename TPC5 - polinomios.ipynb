```python
def menu():
    print("Polinomios")
    print("(1) Criar polinomio")
    print("(2) Calcular polinomio")
    print("(3) Calcular polinomios - tabela")
    print("(4) Simplificar polinomio")
    print("(5) Derivar um polinomio")
    print("(0) Sair")
    
    op = input("Qual e a sua opcao: ")
    
    if op == "1":
        verPolinomio(criarPolinomio())
        
    if op == "2":
        p = criarPolinomio()
        x = int(input("valor de x: "))
        print(calcPol(p,x))
        
    if op == "3":
        p = criarPolinomio()
        verPolinomio(tabela(p))
        
    if op == "4":
        p = criarPolinomio()
        verPolinomio(SimpPol(p))
        
    if op == "5":
        p = criarPolinomio()
        verPolinomio(derivarPol(p))
        
    
def criarTermo(coef,exp):
    
    return (coef,exp)

def criarPolinomio():
    polinomio = []
    print("Criar polinomio - escreve 'stop' para terminar de escrever o polinomio")
    grau = input("Grau do polinomio: ")
    coef = input("coeficiente: " )
    if grau == "stop" or coef == "stop":
            return polinomio
    else:
        grau = int(grau)
        coef = int(coef) 
        termo = criarTermo(coef, grau)
        inserirTermo(polinomio, termo)
        
        while grau != "stop" or coef != "stop" :
            grau = input("Grau do polinomio: ")
            if grau == "stop":
                return polinomio
            
            coef = input("coeficiente: " )
            
            if coef == "stop":
                return polinomio
            
            else:
                grau = int(grau)
                coef = int(coef)       

                termo = criarTermo(coef, grau)
                inserirTermo(polinomio, termo)


def inserirTermo(p,t):
    return p.append(t)



def verTermo(t):
    coef,exp=t
    if exp==0:
        return str(coef)
    else:
        return str(coef)+"x^"+str(exp)
    
def verPolinomio(p):
    res=""
    for t in p:
        if res=="":
            res=verTermo(t)
        else:
            res=res + "+" + verTermo(t)
    print(res)
    

def calcPol(p,x):
    res=0

    for i in  range(len(p)):
        t=(x**p[i][1])*p[i][0]
        res=res+t
    return res

def tabela(p):
    n = int(input("Calcular resultados atÃ© que valor de x? "))
    for i in range (n):
        print("x=",i,", polinomio= ", calcPol(p, i))



def chaveOrd(p):
    _,e=p
    return e
    
def ordenarPolinomio(p):
    p=p.sort(reverse=True, key=chaveOrd)    
    return p
    
def SimpPol(p):
    pol = []
    
    for i in range(len(p)-1):
         if p[i][1] == p[i + 1][1]:
            coef , exp = p[i][0] + p[i+1][0] , p[i][1]
            t = criarTermo(coef, exp)
            inserirTermo(pol, t)
    return pol
    
    
def derivarPol(p):
    res= []
    
    for (c,e) in p:
        if e>=0:
            res.append((c*e,e-1))
    return res



menu()
```
