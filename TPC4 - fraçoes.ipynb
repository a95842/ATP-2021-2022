```python
def menu():
    print("Frações")
    print("(1) Criar fração")
    print("(2) Simplificar fração")
    print("(3) Somar frações")
    print("(4) Subtrair frações")
    print("(5) Multiplicar frações")
    print("(6) Dividir frações")
    print("(7) Criar uma lista de n frações aleatórias")
    print("(8) Somar lista")
    print("(9) Maior elemento da lista")
    print("(0) Sair")
    
    op = input("Qual e a sua opcao: ")
    
    
    if op == "0" :
        exit()
        
    elif op == "1":
        CriarFracao()
        
    elif op == "2":
        num = int(input("numerador da fração desejada: "))
        denom = int(input("denominador da fraçao desejada: "))
        f = CriarFracao(num, denom)
        VerFuncao(SimplificarFracao(f))
        
    elif op == "3":
        num1 = int(input("numerador da primeira fração desejada: "))
        denom1 = int(input("denominador da primeira fraçao desejada: "))
        num2 = int(input("numerador da segunda fração desejada: "))
        denom2 = int(input("denominador da segunda fraçao desejada: "))
        f1 = CriarFracao(num1, denom1)
        f2 = CriarFracao(num2, denom2)
        VerFuncao(SomaFracao(f1,f2))
        
    elif op == "4":
        num1 = int(input("numerador da primeira fração desejada: "))
        denom1 = int(input("denominador da primeira fraçao desejada: "))
        num2 = int(input("numerador da segunda fração desejada: "))
        denom2 = int(input("denominador da segunda fraçao desejada: "))
        f1 = CriarFracao(num1, denom1)
        f2 = CriarFracao(num2, denom2)
        VerFuncao(SubFracao())
        
    elif op == "5":
        num1 = int(input("numerador da primeira fração desejada: "))
        denom1 = int(input("denominador da primeira fraçao desejada: "))
        num2 = int(input("numerador da segunda fração desejada: "))
        denom2 = int(input("denominador da segunda fraçao desejada: "))
        f1 = CriarFracao(num1, denom1)
        f2 = CriarFracao(num2, denom2)
        VerFuncao(MultFracao(f1,f2))
        
    elif op == "6":
        num1 = int(input("numerador da primeira fração desejada: "))
        denom1 = int(input("denominador da primeira fraçao desejada: "))
        num2 = int(input("numerador da segunda fração desejada: "))
        denom2 = int(input("denominador da segunda fraçao desejada: "))
        f1 = CriarFracao(num1, denom1)
        f2 = CriarFracao(num2, denom2)
        VerFuncao(DivFracao(f1,f2))
        
    elif op == "7":
        print(ListaFracoes())
        
    elif op == "8":
        lista = ListaFracoes()
        VerFuncao(SomaLista(lista))
    
    elif op == "9":
        lista = ListaFracoes()
        VerFuncao(MaxLista(lista))
        
        
    else:
        print("Nao existe essa opcao")
        menu()
        
def VerFuncao(f):
    print(str(int(f[0])) + "/" + str(int(f[1])))
    
def CriarFracao(num,denom):
    return(num,denom)


def mdc(a,b):
    if a % b == 0:
        return b
    elif b%a == 0:
        return a
    else:
        if a > b:
            return mdc(b, a%b)
        else:
            return mdc (a, b%a)
        
def SimplificarFracao(f):
    num , denom = f
    a = mdc (int(num), int(denom))
    return(num/a , denom/a)
    

def SomaFracao(f1,f2):
    num1, denom1 = f1
    num2, denom2 = f2
    
    if denom1 == denom2 :
        return(SimplificarFracao(CriarFracao(num1 + num2 , denom1)))
    
    else:
        return(SimplificarFracao(CriarFracao(num1 *denom2 + num2 * denom1, denom1 * denom2)))
    


def SubFracao(f1,f2):
    num1, denom1 = f1
    num2, denom2 = f2
    
    if denom1 == denom2 :
        return(SimplificarFracao(CriarFracao(num1 - num2 , denom1)))
    
    else:
        return(SimplificarFracao(CriarFracao(num1 *denom2 - num2 * denom1, denom1 * denom2)))
    
    
def MultFracao(f1,f2):
    num1, denom1 = f1
    num2, denom2 = f2
    
    
    return(SimplificarFracao(CriarFracao(num1 * num2 , denom1 * denom2)))
    

def DivFracao(f1,f2):
    num1, denom1 = f1
    num2, denom2 = f2
    
    
    return(SimplificarFracao(CriarFracao(num1 * denom2 , denom1 * num2)))
    

def ListaFracoes():
    n = int(input("Número de frações que a lista deverá ter: "))
    lista = []
    i = 1
    import random
    while i != n:
        num = random.randint(0,50)
        denom = random.randint(0,50)
        lista.append(CriarFracao(num, denom))
        i = i +1
        
    return lista


def SomaLista(lista):
    res = (0,1)
    for fracao in lista:
        res = SomaFracao(res, fracao)
        
    return res

def MaxLista(lista):
    irracionais = []
    
    for fracao in lista:
        num, denom = fracao
        irracionais.append(num/denom)
    maximo = irracionais.index(max(irracionais))
    
    return lista[maximo]
    
        
menu()
```
